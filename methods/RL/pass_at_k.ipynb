```python
import pandas as pd
import os
from pathlib import Path
# Qwen2.5-1.5B-Instruct_countdown2345_grpo_gaussian_0.5_0.5_True_1600
test_set = [f'countdown_n{i}t100_1-100' for i in range(2, 7)]
def draw_pass_at_k(model_name):
    # pass_at_k_file = Path(os.getcwd()).parents[1] / 'outputs' / model_name / f"pass_at_k_results_{test_set[-2]}.csv"
    pass_at_k_file = Path('/mnt/data/shared/shurui.gui/sys2bench/') / model_name / f"pass_at_k_results_{test_set[-2]}.csv"
    df_pass = pd.read_csv(pass_at_k_file).transpose()
    df_pass_at_k = df_pass.cummax()
    df_mean_pass_at_k = df_pass_at_k.mean(axis=1)
    df_mean_pass_at_k.plot(y="pass_at_k", ylabel="Pass@k", xlabel="k", title="Avg Pass@k")
# draw_pass_at_k("Qwen2.5-1.5B-Instruct_countdown2345_grpo_gaussian_0.5_0.5_True_1600")
# draw_pass_at_k("Qwen2.5-1.5B-Instruct")
draw_pass_at_k("Llama-3.2-3B-Instruct_countdown2345_grpo_gaussian_0.25_0.75_True_1600")
draw_pass_at_k("Llama-3.2-3B-Instruct")
```
